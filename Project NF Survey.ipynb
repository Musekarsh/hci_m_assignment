```python
%matplotlib inline
import numpy as np
import pandas as pd
#import seaborn as sns
import matplotlib.pyplot as plt
#sns.set()
import os 
cwd = os.getcwd()
print("CWD is {}".format(cwd))
df = pd.read_csv("response.csv")
print(df.shape)
df.columns
```

```python
#Format# Data
del df['response']
df.head()
df.rename(columns={"Q1": "Student", 
                   "Q2": "Age",
                   "Q3": "User", 
                   "Q4": "UsageFreq", 
                   "Q5": "MostUsedFeature",
                   "Q6": "OtherMostUsed",
                   "Q7": "WorksWell",
                   "Q8": "OtherWorksWell",
                   "Q9": "LongDocFreq",
                   "Q10": "LongDocPage",
                   "Q11": "MostImprov",
                   "Q12": "OtherMostImprov",
                   "Q13": "ViewFreq",
                   "Q14": "ViewImprove",
                   "Q15": "NavigateFreq",
                   "Q16": "NavigateImprove",
                   "Q17": "SearchFreq",
                   "Q18": "SearchMethod",
                   "Q19": "SearchSuccess",
                   "Q20": "SearchImprove",
                   "Q21": "AnnotateFreq",
                   "Q22": "AnnotateMethod",
                   "Q23": "AnnotateFlex",
                   "Q24": "AnnotateImprove",
                   "Q25": "HighestPriChange",
                   "Q26": "OtherImprove"},inplace=True)
df.head()
```

```python
value_counts={}
for k in df.columns:
    vc=df[k].value_counts().sort_index(ascending=True)
    value_counts[k]={}
    value_counts[k]['series']=vc
    value_counts[k]['dict']=vc.to_dict()
df.columns
```

```python
fig,axes=plt.subplots(nrows=1, ncols=6,figsize=(15,5))
i=0
for col in ['Student','Age','User','UsageFreq','LongDocFreq','LongDocPage']:
    value_counts[col]['series'].plot.bar(ax=axes[i],title=col)
    i+=1
```

```python
fig,axes=plt.subplots(nrows=1, ncols=3,figsize=(15,5))
i=0
for col in ['MostUsedFeature','WorksWell','MostImprov']:
    value_counts[col]['series'].plot.bar(ax=axes[i],title=col)
    i+=1
```

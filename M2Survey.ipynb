```python
%matplotlib inline
import numpy as np
import pandas as pd
import os 
cwd = os.getcwd()
print("CWD is {}".format(cwd))
df = pd.read_csv("response.csv")
print(df.shape)
df.columns
```

```python
#Format# Data
del df['response']
df.head()
df.rename(columns={"Q1": "Age", 
                   "Q2": "CurrStatus",
                   "Q3": "OwnKindle", 
                   "Q4": "ReadFreq", 
                   "Q5": "KindleTime",
                   "Q6": "Proficiency",
                   "Q7": "CarryIt",
                   "Q9": "NumFilesCollect",
                   "Q10": "TransitionTime",
                   "Q11": "AllocationBy",
                   "Q12": "FilesPerDay",
                   "Q14": "SplitView",
                   "Q15": "IncompleteSplitView",
                   "Q16": "VerticalSplitViews",
                   "Q17": "AllIconViews",
                   "Q18": "AllListViews"},inplace=True)
df.head()
```

```python
value_counts={}
for k in df.columns:
    vc=df[k].value_counts().sort_index(ascending=True)
    value_counts[k]={}
    value_counts[k]['series']=vc
    value_counts[k]['dict']=vc.to_dict()
#print(value_counts['Q1']['dict']['30 - 39'])
#Bar plot for
#Q1: Age
#Q2: Current Status
#Q3: Own Kindle
#Q4: Frequency of reading
#Q5: Kindle Time
#Q6: Proficiency
#Q7: Carry it?
#Q9: Num Files Collections
#Q10: Time b/f transition
#Q11: Allocation By
#Q12: Files opened in a day
#Q14: SplitView
#Q15: Incomplete SplitView
#Q16: Veritcal SplitView
#Q17: All icons
#Q18: All List

#Q8: Subj: Carry where?
#Q13: Subj: First impressions
#Q19: Subj: Change in impressions
#Q20: Subj: E.g. where all do you read?
#Q21: Subj: Redesign
df.columns
df_graph=df[['Age', 'CurrStatus', 'OwnKindle', 'ReadFreq', 'KindleTime',
       'Proficiency', 'CarryIt', 'NumFilesCollect', 'TransitionTime',
       'AllocationBy', 'FilesPerDay', 'SplitView',
       'IncompleteSplitView', 'VerticalSplitViews', 'AllIconViews',
       'AllListViews']]
```
